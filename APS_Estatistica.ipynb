```python
# Importando o pandas
import pandas as pd


def showGraphs(option):
    # Importando a base cardekho
    cars = pd.read_csv('cardekho.csv')
    graphOption = {
        1: cars.fuel.value_counts(),
        2: cars.transmission.value_counts(),
        3: cars.owner.value_counts(),
        4: cars.seller_type.value_counts(),
        5: cars.seats.value_counts(),
        6: cars.groupby('year')['selling_price'].max().sort_values(ascending=False),
        7: cars.groupby('name')['km_driven'].min().sort_values(ascending=True).head(20),
    }.get(option,False)

    graphOption.plot.bar()

#main
print("Digite a opcao que deseja ver o grafico: ")
print("1: Tipo de combustivel")
print("2: Tipo de cambio")
print("3: Donos do carro")
print("4: Tipo de vendedor")
print("5: Quantidade de bancos")
print("6: Agrupado por ano, preco de venda")
print("7: Menor KM")
showGraphs(int(input()))














```
